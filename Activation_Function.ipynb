```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Define the activation functions
def linear(x):
    return x

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

# Create a range of values for the x-axis
x = np.linspace(-10, 10, 100)

# Calculate the activation outputs
y_linear = linear(x)
y_sigmoid = sigmoid(x)
y_tanh = tanh(x)
y_relu = relu(x)
y_leaky_relu = leaky_relu(x)

# Plot the activation functions
plt.figure(figsize=(12, 8))
plt.plot(x, y_linear, label='Linear', linewidth=2)
plt.plot(x, y_sigmoid, label='Sigmoid', linewidth=2)
plt.plot(x, y_tanh, label='Tanh', linewidth=2)
plt.plot(x, y_relu, label='ReLU', linewidth=2)
plt.plot(x, y_leaky_relu, label='Leaky ReLU', linewidth=2)

# Add labels, title, and legend
plt.xlabel('x')
plt.ylabel('Activation output')
plt.title('Activation Functions in Neural Networks')
plt.legend()
plt.grid(True)
plt.show()
```

Activation functions play a critical role in neural networks by introducing non-linearities into the model, allowing it to learn complex patterns and make accurate predictions. Here's a detailed overview of activation functions in neural networks:

1. Why Activation Functions Are Needed
Linear Models: If a neural network only used linear transformations (like matrix multiplication), the output would also be linear, regardless of the depth or complexity of the network. Linear models are inherently limited in their capacity to represent complex relationships.
Non-Linear Models: Activation functions introduce non-linearity, allowing neural networks to model intricate patterns, making them powerful for tasks like image recognition, speech processing, and more.

3. How to Choose an Activation Function
The choice of activation function depends on the problem domain and the network architecture:

Deep Networks: ReLU and its variants (Leaky ReLU, Parametric ReLU, etc.) are common due to their computational efficiency and ability to mitigate vanishing gradients.
Binary Classification: Sigmoid is a common choice because it outputs values between 0 and 1.
Multi-Class Classification: Softmax is usually used for its ability to generate a probability distribution.
Hidden Layers: ReLU and Tanh are common choices due to their non-linearity and performance.
4. Recent Developments
Research into activation functions continues, with new variants proposed regularly. The goal is to find functions that combine non-linearity, differentiability, and computational efficiency while avoiding common pitfalls like vanishing or exploding gradients.

Overall, the choice of activation function is a crucial consideration in neural network design and can significantly impact the performance and training dynamics of the network.
